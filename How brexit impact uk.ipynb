# Earnings
The provided dataframe categorizes information retrieved from a sheet named "People" into four regions: United Kingdom, England, Wales, Scotland, and Northern Ireland. All the earnings within this dataframe are represented with decimal places. I applied the df.round() method to round the data to zero decimal places and utilized df.fillna() to substitute any NaN values with zeros.

```python
# Read data 
df= pd.read_excel('Earnings-region.xls', sheet_name = 'People',index_col=0, skiprows=16, usecols='A,B,D,N,O,P')
df.columns = ['United Kingdom','England','Wales','Scotland','NorthernIreland']

# Replace non-numeric values with NaN
df = df.apply(pd.to_numeric, errors='coerce')

# Replace NaN values with 0
df = df.fillna(0)

# Round up all the data to 0 decimal places
df = df.round(0)

# Convert the DataFrame to integers
df = df.astype(int)

# Only the first 100 rows
df = df.head(98)

df
```

The dataset delineates the total and regional incomes (mean average pounds per week) in the United Kingdom from 2010 to 2021, revealing that England consistently held the highest incomes, trailed by Scotland, Wales, and Northern Ireland. Throughout the Brexit phase, all regions experienced a positive trajectory with increasing incomes. However, in the post-Brexit periods(after 2020), a temporary decline in incomes across all areas occurred for two consecutive quarters, followed by a resurgence in 2021. Notably, England stood out with a continuous upward trend in incomes during this phase, underscoring its economic resilience amidst broader transitional shifts.Brexit has influenced the income dynamics in the United Kingdom, as reflected in a temporary decline in incomes across all regions post-2020, followed by a resurgence, highlighting the economic impact of broader transitional shifts during this period.

```python
# Filter rows from the year 2010 onwards
df = df[df.index.str.contains('2010') | df.index.str.contains('2011') | df.index.str.contains('2012') | df.index.str.contains('2013') | df.index.str.contains('2014') | df.index.str.contains('2015') | df.index.str.contains('2016') | df.index.str.contains('2017') | df.index.str.contains('2018') | df.index.str.contains('2019') | df.index.str.contains('2020') | df.index.str.contains('2021')]

# Plot the line graph
plt.figure(figsize=(15, 6)) 

# Plot each column as a separate line
for column in df.columns:
    plt.plot(df.index, df[column], label=column)
    
plt.xlabel('Time')
plt.ylabel('Earnings')
plt.title('Earnings by Region Over Time (2010-2021)')
plt.legend()  
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

The table presented below provides a summary of key economic indicators related to earnings, covering the periods of pre-Brexit, Brexit, and post-Brexit.

```python
# Extract the rows corresponding to (Pre-Brexit)
pre_brexit_data = df.loc["Oct-Dec 2016"]

# Extract the rows corresponding to (Brexit)
brexit_data = df.loc["Jan-Mar 2020 (r)"]

# Extract the rows corresponding to (Post-Brexit)
post_brexit_data = df.loc["Jul-Sep 2020 (r)"]

# Create a new DataFrame with the extracted data
selected_periods_df = pd.concat([pre_brexit_data, brexit_data, post_brexit_data], axis=1)
selected_periods_df.columns = ["Pre-Brexit", "Brexit", "Post-Brexit"]

# Transpose the DataFrame
selected_periods_df = selected_periods_df.T
selected_periods_df.index.name = 'Periods'
selected_periods_df.columns.name = 'The Region'

selected_periods_df
```

This analysis underscores the complexity of the economic impact of Brexit on different regions within the United Kingdom. It provides an overview of the average regional earnings for the Pre-Brexit, Brexit and Post-Brexit periods. We can see that the overall trend is positive, regional disparities exist. England remains a key economic driver, showcasing consistent growth. Scotland demonstrates substantial growth, possibly influenced by regional factors. Wales experienced a temporary decline but rebounded in the post-Brexit period. Northern Ireland's modest increase reflects a stable trajectory.

```python
# Plotting the grouped bar chart
fig, ax = plt.subplots(figsize=(10, 6))

bar_positions = np.arange(len(selected_periods_df.index))
bar_width = 0.1

# Plotting bars for each region with space between periods
for i, region in enumerate(selected_periods_df.columns):
    ax.bar(bar_positions + i * bar_width, selected_periods_df[region], width=bar_width, label=region)

ax.set_xlabel('Periods')
ax.set_ylabel('Total Earnings')
ax.set_title('Total Earnings by Period for Each Region')
ax.set_xticks(bar_positions + (len(selected_periods_df.columns) - 1) * bar_width / 2)
ax.set_xticklabels(selected_periods_df.index)
ax.legend()

# Display the plot
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')

# Filter data for December 2020
dec_2020 = df[(df['Date'] == '2020-12-01')]

# Filter data for January 2021
jan_2021 = df[(df['Date'] == '2021-01-01')]

# Merge the two DataFrames on 'Category'
merged_df = pd.merge(dec_2020, jan_2021, on='Category', how='outer', suffixes=('_Dec_2020', '_Jan_2021'))

# Calculate the difference in exports between December 2020 and January 2021
merged_df['Exports_Difference'] = merged_df['Exports_Jan_2021'] - merged_df['Exports_Dec_2020']

# Display the result
print("Exports Difference between December 2020 and January 2021:")
print(merged_df[['Category', 'Exports_Dec_2020', 'Exports_Jan_2021', 'Exports_Difference']])
```

The significant decrease in exports, as indicated by negative differences, suggests disruptions in supply chains and changes in trade patterns. These shifts align with the challenges posed by new trade regulations and border controls post-Brexit. # Migration
The post-Brexit immigration rules exempt EU citizens and their eligible family members residing in the UK before December 31, 2020, from the new regulations. They needed to apply to the EU Settlement Scheme (EUSS) for pre-settled or settled status. Settled status is a form of permanent residency, required for UK citizenship, while pre-settled status is for those in the UK for under 5 years. Initially set to expire after 5 years, pre-settled status will now no longer expire due to a 2022 court decision. The EUSS application deadline was June 30, 2021, with late applications permitted under reasonable grounds. Newly arriving EU citizens must obtain a visa for work, family, or study, with skilled workers requiring a job offer for a middle or high-skilled role. For family reunification, an £18,600 income threshold applies, and EU students now face higher tuition fees and must apply for a student visa post-free movement[5].

```python
df = pd.read_excel("Immigration.xlsx",index_col=0, skiprows=6)
df.columns = ['Non-EU','EU','British','Total long-Term Immigration']

df
```

This decline suggests a potential impact of Brexit-related policies or economic conditions on the decision of EU nationals to migrate to the UK. Concurrently, Non-EU immigration displays a more gradual decrease over the same period. In contrast, British immigration remains relatively stable throughout the timeline. # Word Count

The following code will count the number of words in Markdown cells. Code cells are not included.

- `Main word count` is the number of words in the main body of the text, *excluding* references or appendices.
- `References and appendices word count` is the number of words in any references or appendices.

Only `Main word count` relates to the assignment word limit. There is no limit to the number of words that can be included in references or appendices. Please note that appendices should only be used to provide context or supporting information. *No marks will be directly awarded for material submitted in appendices*.

Important:

- Please do not modify the word count code!
- To exclude references from your word count **you must** have a cell that starts with the text `## References`. Everything below this cell will not count towards the main word count.
- If you are submitting additional material as appendices **you must** have a cell that starts with the text `## Appendices`. Everything below this cell will not count towards the main word count. If you do not have any appendices you can delete the `## Appendices` cell.
- Code comments should only be used to explain details of the implementation, not for discussing your findings. All analysis commentary **must** be written in Markdown cells. *No marks will be awarded for analysis discussion submitted as comments in code cells*.

```python
%%js

// Run this cell to update your word count.

function wordcount() {
    let wordCount = 0
    let extraCount = 0
    let mainBody = true

    let cells = Jupyter.notebook.get_cells()
    cells.forEach((cell) => {
        if (cell.cell_type == 'markdown') {
            let text = cell.get_text()
            // Stop counting as main body when get to References or Appendices.
            if (text.startsWith('## References') ||
                text.startsWith('## Appendices')) {
                mainBody = false
            }
            if (text.startsWith('## Word Count')) {
                text = ''
            }
            if (text) {
                let words = text.toLowerCase().match(/\b[a-z\d]+\b/g)
                if (words) {
                    let cellCount = words.length
                    if (mainBody) {
                        wordCount += cellCount
                    } else {
                        extraCount += cellCount
                    }
                }
            }
        }
    })
    return [wordCount, extraCount]
}

let wc = wordcount()
element.append(`Main word count: ${wc[0]} (References and appendices word count: ${wc[1]})`)
```
